```python
#Do some preprocessing before correlation matrix

#Drop columns with 1 unique value
df = df.drop(df.columns[df.nunique() == 1], axis=1)
#Drop Nan
df = df.dropna()


drop_cols = ['Date', 'Timestamp', 'Scr_IP', 'Scr_port', 'Des_IP', 'Des_port',
       'Protocol']
df = df.drop(drop_cols, axis=1)

print(len(df.columns))
```

```python
#Encode objects as ints
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

#convert where Duration = "-" into 0
df['Duration'] = df['Duration'].replace('-', '0')
df['Duration'] = df['Duration'].astype(float)

df['Scr_bytes'] = df['Scr_bytes'].replace('-', '0')
df['Scr_bytes'] = df['Scr_bytes'].astype(float)

df['Des_bytes'] = df['Des_bytes'].replace('-', '0')
df['Des_bytes'] = df['Des_bytes'].astype(float)

df['missed_bytes'] = df['missed_bytes'].replace('-', '0')
df['missed_bytes'] = df['missed_bytes'].astype(float)

df['Scr_pkts'] = df['Scr_pkts'].replace('-', '0')
df['Scr_pkts'] = df['Scr_pkts'].astype(float)

df['Scr_ip_bytes'] = df['Scr_ip_bytes'].replace('-', '0')
df['Scr_ip_bytes'] = df['Scr_ip_bytes'].replace('excel', '0')  #1 row contains excel
df['Scr_ip_bytes'] = df['Scr_ip_bytes'].astype(float)

df['Des_pkts'] = df['Des_pkts'].replace('-', '0')
df['Des_pkts'] = df['Des_pkts'].astype(float)

df['Des_ip_bytes'] = df['Des_ip_bytes'].replace('-', '0')
df['Des_ip_bytes'] = df['Des_ip_bytes'].astype(float)

df["anomaly_alert"] = df["anomaly_alert"].replace('-', 'FALSE')

df["total_bytes"] = df["total_bytes"].replace('-', '0')
df["total_bytes"] = df["total_bytes"].astype(float)

df["total_packet"] = df["total_packet"].replace('-', '0')
df["total_packet"] = df["total_packet"].astype(float)

df["paket_rate"] = df["paket_rate"].replace('-', '0')
df["paket_rate"] = df["paket_rate"].astype(float)

df["byte_rate"] = df["byte_rate"].replace('-', '0')
df["byte_rate"] = df["byte_rate"].replace('?', '0')
df["byte_rate"] = df["byte_rate"].astype(float)

df["Scr_packts_ratio"] = df["Scr_packts_ratio"].replace('-', '0')
df["Scr_packts_ratio"] = df["Scr_packts_ratio"].astype(float)

df["Des_pkts_ratio"] = df["Des_pkts_ratio"].replace('-', '0')
df["Des_pkts_ratio"] = df["Des_pkts_ratio"].astype(float)

df["Scr_bytes_ratio"] = df["Scr_bytes_ratio"].replace('-', '0')
df["Scr_bytes_ratio"] = df["Scr_bytes_ratio"].astype(float)

df["Des_bytes_ratio"] = df["Des_bytes_ratio"].replace('-', '0')
df["Des_bytes_ratio"] = df["Des_bytes_ratio"].astype(float)

df['Avg_ideal_time'] = df['Avg_ideal_time'].replace('-', '0')   
df['Avg_ideal_time'] = df['Avg_ideal_time'].astype(float)

df['Avg_user_time'] = df['Avg_user_time'].replace('-', '0')
df['Avg_user_time'] = df['Avg_user_time'].replace('aza', '0')
df['Avg_user_time'] = df['Avg_user_time'].replace('?', '0')
df['Avg_user_time'] = df['Avg_user_time'].replace(' ', '0')
df['Avg_user_time'] = df['Avg_user_time'].astype(float)

df['Avg_nice_time'] = df['Avg_nice_time'].replace('-', '0')
df['Avg_nice_time'] = df['Avg_nice_time'].astype(float)

df['Avg_system_time'] = df['Avg_system_time'].replace('-', '0')
df['Avg_system_time'] = df['Avg_system_time'].replace('?', '0')
df['Avg_system_time'] = df['Avg_system_time'].astype(float)

df['Avg_iowait_time'] = df['Avg_iowait_time'].replace('-', '0')
df['Avg_iowait_time'] = df['Avg_iowait_time'].replace('?', '0')
df['Avg_iowait_time'] = df['Avg_iowait_time'].astype(float)

df['Avg_tps'] = df['Avg_tps'].replace('-', '0')
df['Avg_tps'] = df['Avg_tps'].astype(float)

df['Avg_rtps'] = df['Avg_rtps'].replace('-', '0')
df['Avg_rtps'] = df['Avg_rtps'].astype(float)

df['Avg_wtps'] = df['Avg_wtps'].replace('-', '0')
df['Avg_wtps'] = df['Avg_wtps'].astype(float)

df['Avg_ldavg_1'] = df['Avg_ldavg_1'].replace('-', '0')
df['Avg_ldavg_1'] = df['Avg_ldavg_1'].astype(float)

df['Std_ideal_time'] = df['Std_ideal_time'].replace('-', '0')
df['Std_ideal_time'] = df['Std_ideal_time'].astype(float)

df['Std_user_time'] = df['Std_user_time'].replace('-', '0')
df['Std_user_time'] = df['Std_user_time'].astype(float)

df['Std_nice_time'] = df['Std_nice_time'].replace('-', '0')
df['Std_nice_time'] = df['Std_nice_time'].astype(float)

df['Std_system_time'] = df['Std_system_time'].replace('-', '0')
df['Std_system_time'] = df['Std_system_time'].replace('?', '0')
df['Std_system_time'] = df['Std_system_time'].astype(float)

df['Std_iowait_time'] = df['Std_iowait_time'].replace('-', '0')
df['Std_iowait_time'] = df['Std_iowait_time'].replace('?', '0')
df['Std_iowait_time'] = df['Std_iowait_time'].astype(float)

df['Std_tps'] = df['Std_tps'].replace('-', '0')
df['Std_tps'] = df['Std_tps'].astype(float)

df['Std_rtps'] = df['Std_rtps'].replace('-', '0')
df['Std_rtps'] = df['Std_rtps'].astype(float)

df['Std_wtps'] = df['Std_wtps'].replace('-', '0')
df['Std_wtps'] = df['Std_wtps'].astype(float)

df['Std_ldavg_1'] = df['Std_ldavg_1'].replace('-', '0')
df['Std_ldavg_1'] = df['Std_ldavg_1'].astype(float)

df['Avg_kbmemused'] = df['Avg_kbmemused'].replace('-', '0')
df['Avg_kbmemused'] = df['Avg_kbmemused'].replace('#DIV/0!', '0')
df['Avg_kbmemused'] = df['Avg_kbmemused'].replace('?', '0')
df['Avg_kbmemused'] = df['Avg_kbmemused'].astype(float)

df['Std_kbmemused'] = df['Std_kbmemused'].replace('-', '0')
df['Std_kbmemused'] = df['Std_kbmemused'].replace('#DIV/0!', '0')
df['Std_kbmemused'] = df['Std_kbmemused'].replace('?', '0')
df['Std_kbmemused'] = df['Std_kbmemused'].astype(float)

df['Avg_num_Proc/s'] = df['Avg_num_Proc/s'].replace('-', '0')
df['Avg_num_Proc/s'] = df['Avg_num_Proc/s'].astype(float)

df['Std_num_proc/s'] = df['Std_num_proc/s'].replace('-', '0')
df['Std_num_proc/s'] = df['Std_num_proc/s'].astype(float)

df['Avg_num_cswch/s'] = df['Avg_num_cswch/s'].replace('-', '0')
df['Avg_num_cswch/s'] = df['Avg_num_cswch/s'].astype(float)

df['std_num_cswch/s'] = df['std_num_cswch/s'].replace('-', '0')
df['std_num_cswch/s'] = df['std_num_cswch/s'].astype(float)


for col in df.columns:
    if df[col].dtype == 'object':
        #Attempt to convert to float
        try:
            df[col] = df[col].astype(float)
        except:
            print(df[col].value_counts())
            #If it fails, encode as int
            df[col] = le.fit_transform(df[col])

#Convert to float
df = df.astype(float)
```

```python
corr = df.corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
#decrease font size
sns.set(font_scale=0.5)
sns.heatmap(corr, annot=True, fmt=".1f")
plt.show()

#Get the features that are highly correlated with the target
print(corr["class1"].abs().sort_values(ascending=False))

print(corr["class3"].abs().sort_values(ascending=False))


#Get the labels of most correlated features in matrix
corr_pairs = corr.unstack()
top_correlated_pairs = corr_pairs.abs().sort_values(ascending=False)
top_correlated_pairs = top_correlated_pairs[
    (top_correlated_pairs.index.get_level_values(0) != top_correlated_pairs.index.get_level_values(1))
]
# Filter to keep only unique pairs (ignore mirror entries and self-correlation)
top_correlated_pairs = top_correlated_pairs[
    (top_correlated_pairs.index.get_level_values(0) != top_correlated_pairs.index.get_level_values(1))
]

```
